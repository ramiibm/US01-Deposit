Crossed validation to evaluate the true performance of the classifier that is learned using the whole data

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import ShuffleSplit
clf = tree.DecisionTreeClassifier(max_depth=9)
scores = cross_val_score(clf, X, Z, cv=ShuffleSplit(n_splits=5))
np.mean(scores)
```

Visualize the decision tree to know the used rules

```python
clf = tree.DecisionTreeClassifier(max_depth=2)
clf.fit(X,Z)
tree.plot_tree(clf)
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {'max_depth': [2, 4, 6, 8, 10, 20, 40], 'criterion': ['gini','entropy']}
search = GridSearchCV(tree.DecisionTreeClassifier(), param_grid, cv=ShuffleSplit(n_splits=5))
search.fit(X,Z)
```
