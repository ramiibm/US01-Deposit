```python
data['country_macro']['term']['area'].items()
```

```python
def get_country_code(country_name):
    for key, value in data['country_macro']['term']['area'].items():
        code = key
        if value['full'] == country_name or value['short'] == country_name:
            return(code)
    return None
```

```python
code = get_country_code('France')
print(code)
```

```python
for indicator_name, indicator_code in indicator_keys.items():
    print('######### ' + indicator_name + ' ' + indicator_code + ' #########')
    for index, name in myfile['Country (according to CN Gov):'].items():
        country_key = get_country_code(name)
        print(name)
        if country_key is not None:
            try:
                datum = float(data['country_macro']['data'][year][country_key][indicator_code])
                myfile[indicator_name].loc[index] = datum
            except ValueError:
                print('error with ', name, indicator_name, data['country_macro']['data'][year][country_key][indicator_code])
            except KeyError:
                print('key error with', indicator_name)
```
