```python
iso_codes = {}
```

```python
for i in myfile.index:
    name = myfile['Country (according to CN Gov):'][i]
    
    if name in iso_codes:
        continue
    
    clean_name = name.split(' (')[0].strip()
    
    if clean_name == 'Thailiand':
        clean_name = 'Thailand'
    elif clean_name == 'The United Arab Emirates':
        clean_name = 'United Arab Emirates'
    
    url = 'https://restcountries.eu/rest/v2/name/' + clean_name
    r = requests.get(url)
    if r.status_code == 200:
        data = r.json()
        iso_codes[name] = data[0]['alpha3Code']
    else:
        print('[ERROR] we cannot get the data, reason is ', r.status_code, r.text)
        print(clean_name)
```

```python
indicator_value = {}
```

```python
import time
```

```python
for i in myfile.index:
    isocode_name = myfile['isocode'][i]
    print(isocode_name)
    
    if isocode_name in indicator_value:
        continue
    time.sleep(5)
    print("time")
    
    if isocode_name == "IOT":
        url = 'http://api.worldbank.org/v2/country/IND/indicator/GC.TAX.IMPT.ZS?format=json'
    else:
        url = 'http://api.worldbank.org/v2/country/' + isocode_name + '/indicator/GC.TAX.IMPT.ZS?format=json'
    
    print("url")
    r = requests.get(url)
    if r.status_code == 200:
        print("200")
        data = r.json()
        date_cursor = 0;
        
        if isocode_name == "BRN" or isocode_name == "LAO" or isocode_name == "MNE" or isocode_name == "OMN" or isocode_name == "PAK" or isocode_name == "QAT" or isocode_name == "SYR" or isocode_name == "TKM" or isocode_name == "VNM" or isocode_name == "YEM":
            indicator_value[isocode_name] = "No value"
        else:
            indicator_value[isocode_name] = data[1][date_cursor]['value']
        while indicator_value[isocode_name] is None:
            date_cursor += 1
            indicator_value[isocode_name] = data[1][date_cursor]['value']
            
    else:
        print('[ERROR] we cannot get the data, reason is ', r.status_code, r.text)
        print(isocode_name)
    
```
