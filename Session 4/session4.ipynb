```python
for index, name in myfile['Country (according to CN Gov):'].iteritems():
    
    clean_name = name.split(' (')[0].strip()
    
    if clean_name == 'Thailiand':
        clean_name = 'Thailand'
    elif clean_name == 'The United Arab Emirates':
        clean_name = 'United Arab Emirates'
    elif clean_name == 'Brunei':
        clean_name = 'Brunei Darussalam'
    elif clean_name == 'Myanmar':
        clean_name = 'Burma'
    elif clean_name == 'East Timor':
        clean_name = 'Timor-Leste'
    elif clean_name == 'Kyrgyzstan':
        clean_name = 'Kyrgyz Republic'
    elif clean_name == 'United States of America':
        clean_name = 'United States'
        
    try:
        ind = heritage_df[heritage_df['Country Name'] == clean_name]['2019 Score']
        rank = heritage_df[heritage_df['Country Name'] == clean_name]['World Rank']
        myfile.at[index, 'HF Index of Economic Freedom (Scale: 1-100)'] = ind
        myfile.at[index, 'Heritage Foundation Index of Economic Freedom (Global Ranking)'] = rank
    except ValueError:
        print('[critical]:no data for', clean_name)
        continue
```

```python
ratings_df = pd.read_csv('../../2019-US01-Data-Science/data-ratings-2015.csv', sep=';')
```

```python
ratings_df.head()
```

```python
import requests
```

```python
iso_codes = {}
for i in myfile.index:
    name = myfile['Country (according to CN Gov):'][i]
    print(name)
    
    if name in iso_codes:
        continue
    
    clean_name = name.split(' (')[0].strip()
    
    if clean_name == 'Thailiand':
        clean_name = 'Thailand'
    elif clean_name == 'The United Arab Emirates':
        clean_name = 'United Arab Emirates'
    
    url = 'https://restcountries.eu/rest/v2/name/' + clean_name
    r = requests.get(url)
    if r.status_code == 200:
        data = r.json()
        iso_codes[name] = data[0]['alpha2Code']
    else:
        print('[ERROR] we cannot get the data, reason is ', r.status_code, r.text)
        print(clean_name)
```

```python
iso_codes
```

```python
myfile['isocode2'] = myfile['Country (according to CN Gov):'].apply(lambda x: iso_codes[x])
```

```python
for index, code in myfile.isocode2.iteritems():
    try:
        sp = ratings_2015[ratings_2015['Country2Code'] == code]['SP_NumericRating']
        myfile.at[index, 'S&P Sovereign Ratings List'] = sp
        
        moody = ratings_2015[ratings_2015['Country2Code'] == code]['Moodys_NumericRating']
        myfile.at[index, "Moody's Socvereign Ratings List"] = moody

        fitch = ratings_2015[ratings_2015['Country2Code'] == code]['Fitch_NumericRating']
        myfile.at[index, 'Fitch_NumericRating'] = fitch
        print(fitch)
    except ValueError:
        print('[critical]:no data for', code)
        continue
```
