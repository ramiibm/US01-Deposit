Importer bibliothèque pandas

```python
import pandas as pd
```

Lire dataset excel. 

```python
pd.read_excel('../Data-sheet-STUDENTS.xlsx')
```

On observe deux majeurs problèmes d'importation : l'en tête du fichier n'étant pas correctement situé, ce dernier est considéré comme étant la première observation de notre dataset ; ensuite, l'index du dataset étant déjà existant, il faut préciser à pandas de l'utiliser et de ne pas en créer un nouveau. On insère également ce dataframe dans un 

```python
myfile = pd.read_excel('../Data-sheet-STUDENTS.xlsx', header=2,index_col=0)
```

On affiche ici le nom du premier pays de notre dataframe.

```python
print(myfile['Country (according to CN Gov):'][1])
```

On affiche ici un à un tous les pays de notre dataframe.

```python
for i in myfile.index:
    print(myfile['Country (according to CN Gov):'][i])
```

Enfin, on y ajoute l'url de l'api nous retournant l'isocode de chacun de ces pays.

```python
for i in myfile.index:
    print('https://restcountries.eu/rest/v2/name/' + myfile['Country (according to CN Gov):'][i])
```
