Display every columns to help us choose the ones we want to keep

```python
clean_df = df[['Country (according to CN Gov):',
       'HF Index of Economic Freedom (Scale: 1-100)',
       'Bertelsmann Transformation index (BTI)',
       'WB UNESCO R&D Expenditure % of GDP',
       'WB UNESCO Researchers p. mn inhabitants',
       'TI Corruption Perception Index',
       'BRI GDP p.c.']]
```

Retirer toutes les lignes n'Ã©tant pas complÃ¨tement remplies

```python
clean_df = clean_df.dropna()
clean_df.head()
```

We want to set the GDP in classes (low gdp, average gdp, high gdp) rather than a raw number because it is easier to predict. To do so, we need to have the distribution of our column.

```python
clean_df['BRI GDP p.c.'].plot()
```

The issue here is that the column is not considered as a number, but as an object (probably because of NaN elements saw earlier). We have to transform it into a float type.

```python
clean_df['BRI GDP p.c.'] = clean_df['BRI GDP p.c.'].apply(float)
clean_df['BRI GDP p.c.'].describe()
```

Plot the distribution

```python
clean_df['BRI GDP p.c.'].plot.hist(bins=20, cumulative=True)
```

```python
clean_df['BRI GDP p.c.'].sort_values().plot(use_index=False)
```

Cut the distribution into 3 intervals : 0 -> 5k for low, 5k -> 20k for avg, 20k -> max for high

```python
clean_df['GDP'] = pd.cut(clean_df['BRI GDP p.c.'], 
       bins=[0,5000,20000,max(clean_df['BRI GDP p.c.'])], 
       labels=['low', 'medium', 'high'])
clean_df.head()
```

We would like to center the distribution around zero (removing the mean) and dividing by the std: centrer rÃ©duire

```python
clean_df['HF Index of Economic Freedom (Scale: 1-100)'].sort_values().plot(use_index=False)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
clean_df['HF_scaled'] = scaler.fit_transform(clean_df[['HF Index of Economic Freedom (Scale: 1-100)']])
clean_df['HF_scaled']
```

```python
clean_df['Bertelsmann_scaled'] = scaler.fit_transform(clean_df[['Bertelsmann Transformation index (BTI)']])
clean_df['R&DExpenditure_scaled'] = scaler.fit_transform(clean_df[['WB UNESCO R&D Expenditure % of GDP']])
clean_df['Researchers_scaled'] = scaler.fit_transform(clean_df[['WB UNESCO Researchers p. mn inhabitants']])
clean_df['TI_scaled'] = scaler.fit_transform(clean_df[['TI Corruption Perception Index']])
```

```python
Export this data into a csv file to process it into rapidminer
```

```python
df_export = clean_df[['HF_scaled', 'Bertelsmann_scaled', 'R&DExpenditure_scaled', 'Researchers_scaled', 'TI_scaled', 'GDP']]
```

```python
df_export.to_csv('gdp_modeling.csv')
```

Now, let's create a model in Python, with *scikit-learn*

Numpy -> great tool to manipulate numbers, matrix...

```python
import numpy as np
from sklearn.ensemble import RandomForestClassifier
```

Select the line randomly: 75% will allow the model to train & 25% will serve to test

```python
df_export['is_train'] = np.random.uniform(0, 1, len(df_export)) < 0.74
df_export
```

Split the df into 2 dfs, one for the train & one for the test

```python
train, test = df_export[df_export['is_train']==True], df_export[df_export['is_train']==False]
```

We know that the first four columns are our features and the fifth (GDP) is the target

```python
features = df_export.columns[0:5]
features
```

Here, we set what we want to achieve

```python
y_train = np.array(train['GDP'])
y_train
```

Here, we create the engine, then fit the model with what we train on and what we want to achieve. Then, we make a prediction, but instead of giviing him the train data, we'll give him the test data. According with what you learned during the trained phase, what will you predict on the test data. Finally, we will compare the two results to see if the predictions are close to the result (send the classification report).

Here, we choose the RandomForestClassifier tuning, but there is a lot of others tunings that would produce better results.

```python
from sklearn.metrics import classification_report

rfc = RandomForestClassifier()

rfc.fit(train[features], y_train)

y_predict = rfc.predict(test[features])
y_true = test['GDP']

print(classification_report(y_true, y_predict))
```
